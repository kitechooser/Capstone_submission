## 1. Project Relevance and Problem Definition

### Current Project Overview
The current codebase implements a machine learning system for detecting fake images, with the following components:
- MesoNet, CNN, SVM or Random Forest architectures for fake image detection
- Bayesian or Optuna optimisation for hyperparameter tuning
- Comprehensive metrics tracking and visualization

### Project Context
This classification project was chosen as I have a background in image processing and it was not only of interest to me, but I am also still working with digital forensics departments in my current role so there is a relevancy for me.

In the main code base there is a selection of ML models to use, SVM, Random Forest, CNN and Mesonet (a lighterweight CNN). This notebook only uses the Mesonet variant though that can be switched here also.

### for best results run main.py from the terminal to see the complete functionality

The choice of Mesonet was made as it used in a reference implementation on the Kaggle site linked above. There are also other papers that discuss the uses these lighterweight CNNs that look at features between whole images and pixels, so texture, edges, etc. Additionally it was the best performing of the four models.

### Other configutation for Optimisation

This shows a segment of the Configuration class that allows the user to refine the default parameters as well as the search space for optimisations. Setting the upper and lower bound identically effectively removes the parameter from the optimisation. 

```python

                    }
                },
                "search_space": [
                    Real(1e-4, 1e-2, name="learning_rate", prior="log-uniform"),
                    Integer(1, 4, name="num_conv_layers"),

                    Categorical(["3x3", "5x5"], name="kernel_size"),

```
Gaussian Process Bayesian Process optimisation was chosen for ease of implementation as well as Optuna as it was brought up on one of the modules previously.

This code segment shows a model being created with parameters from above, metrics being stored during training and the model being fitted

```python
try:
                # Create and train model
                model = self.model_factory.create_model(model_name, params)
                if model is None:
                    print(f"Error creating model {model_name}")
                    print(f"Params: {params}")
                    return 1.0
                
                # Train model
                history = model.fit(
                    self.train_data[0], self.train_data[1],
                    validation_data=(self.val_data[0], self.val_data[1]),
                    epochs=params["training_params"]["epochs"],
                    batch_size=params["training_params"]["batch_size"],
                    verbose=0
                )
                
                # Get validation loss
                val_loss = min(history.history['val_loss'])
                
                # Store trial results
                self.optimization_results["trials"].append({
                    "parameters": param_dict,
                    "metric": float(val_loss)
                })
                
                print(f"Validation Loss: {val_loss:.4f}")
                return float(val_loss)
                
            except Exception as e:
                print(f"Error in objective function: {str(e)}")
                return 1.0

        # Run optimization
        result = gp_minimize(
            func=objective,
            dimensions=search_space,
            n_calls=self.config.OPTIMIZATION_TRIALS,

            noise=1e-10,
            random_state=42
        )
```

## MesoNet with Optuna Optimization
 
### The remainder of this notebook demonstrates using the existing codebase to run MesoNet with Optuna optimization. It is scaled down from runs reported as they are too time consuming to run in the notebook.


### 1. Setup and Imports


````python
## NB. this code has only been tested with Python 3.10 in an Ananconda environment to allow use of Apple GPU 

import sys
import os
import json
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

# Add src directory to path
# sys.path.append(os.path.join(os.getcwd(), '..'))

from src.config.config import Config
from src.models.factory import ModelFactory
from src.training.trainer import ModelTrainer
from src.data.data_loader import DataLoader
````


### 2. GPU Configuration
### Check GPU availability and configure memory growth


````python
try:
    physical_devices = tf.config.list_physical_devices('GPU')
    if physical_devices:
        for device in physical_devices:
            tf.config.experimental.set_memory_growth(device, True)
        print("GPU available and configured")
    else:
        print("No GPU devices found")
except Exception as e:
    print(f"Error configuring GPU: {str(e)}")
    print("Falling back to CPU")
````


### 3. Configuration
 
### Update the configuration to use Optuna optimization




````python
# Create config instance
config = Config()

# Data path
config.BASE_PATH = os.path.join(os.getcwd(), '../image_data/real_and_fake_imageset')

# Set optimization strategy to Optuna
config.OPTIMIZATION_STRATEGY = "Optuna"
config.OPTIMIZATION_TRIALS = 20  # Number of trials to run

# Activate only MesoNet
config.ACTIVE_MODELS = {
    "CNN": False,
    "Mesonet": True,
    "SVM": False,
    "Random Forest": False
}

# disable dashboards for visualization
config.SHOW_DASHBOARDS = False

# Print MesoNet search space
print("\nMesoNet Search Space:")
mesonet_config = config.MODEL_CONFIGS["Mesonet"]
for param in mesonet_config["search_space"]:
    print(f"{param.name}: {param}")
````


### 4. Load Data
### Load the image dataset using the existing data loader


````python

data_loader = DataLoader(config)

# Load and preprocess data
train_data, val_data, test_data = data_loader.load_datasets()

print("\nData shapes:")
print(f"Training: {train_data[0].shape}")
print(f"Validation: {val_data[0].shape}")
print(f"Test: {test_data[0].shape}")

# Optional: Display sample images
plt.figure(figsize=(10, 5))
for i in range(5):
    plt.subplot(1, 5, i+1)
    plt.imshow(train_data[0][i])
    plt.title(f"Label: {train_data[1][i]}")
    plt.axis('off')
plt.tight_layout()
plt.show()
````


### 5. Initialise Model Factory and Trainer


````python
# Create model factory
model_factory = ModelFactory(config)

# Create trainer
trainer = ModelTrainer(config, model_factory)
````


### 6. Train and Evaluate

### Run the training process with Optuna optimisation


````python
# Start training and evaluation
trainer.train_and_evaluate(train_data, val_data, test_data)
````

### 7. Analyse Results
 
### Load and analyse the optimisation results


````python
# Load optimisation results
try:
    with open(os.path.join(config.SAVE_DIR, 'mesonet_stats.json'), 'r') as f:
        stats = json.load(f)
        
    print("\nOptimization Results:")
    print(f"Best metric: {stats['best_metric']:.4f}")
    print(f"Number of trials: {stats['num_trials']}")
    print("\nParameter Importance:")
    for param, importance in stats['parameter_importance'].items():
        print(f"  {param}: {importance:.4f}")
        
except FileNotFoundError:
    print("Stats file not found. Make sure training completed successfully.")
except Exception as e:
    print(f"Error loading results: {str(e)}")
````

### 8. Load Best Model 
### Load and examine the best model found during optimization

````python
try:
    # Load best model
    best_model = tf.keras.models.load_model(
        os.path.join(config.SAVE_DIR, 'best_mesonet.keras')
    )
    
    # Print model summary
    print("\nBest Model Architecture:")
    best_model.summary()
    
    # Load best parameters
    with open(os.path.join(config.SAVE_DIR, 'mesonet_parameters.json'), 'r') as f:
        best_params = json.load(f)
        
    print("\nBest Parameters:")
    print(json.dumps(best_params, indent=2))
    
except FileNotFoundError:
    print("Model or parameter files not found. Make sure training completed successfully.")
except Exception as e:
    print(f"Error loading model: {str(e)}")
````

### 9. Make Predictions

### Use the best model to make predictions on test data

````python
try:
    # Load best model if not already loaded
    if 'best_model' not in locals():
        best_model = tf.keras.models.load_model(
            os.path.join(config.SAVE_DIR, 'best_mesonet.keras')
        )
    
    # Create test dataset
    test_dataset = tf.data.Dataset.from_tensor_slices(test_data[0])\
        .batch(config.BATCH_SIZE)\
        .prefetch(tf.data.AUTOTUNE)
    
    # Make predictions
    predictions = best_model.predict(test_dataset)
    predictions_binary = (predictions > 0.5).astype(int).flatten()
    
    # Calculate metrics
    from sklearn.metrics import classification_report, confusion_matrix
    print("\nClassification Report:")
    print(classification_report(test_data[1], predictions_binary))
    
    # Plot confusion matrix
    cm = confusion_matrix(test_data[1], predictions_binary)
    plt.figure(figsize=(8, 6))
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title('Confusion Matrix')
    plt.colorbar()
    tick_marks = np.arange(2)
    plt.xticks(tick_marks, ['Fake', 'Real'])
    plt.yticks(tick_marks, ['Fake', 'Real'])
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    
    # Add text annotations to confusion matrix
    thresh = cm.max() / 2.
    for i, j in np.ndindex(cm.shape):
        plt.text(j, i, format(cm[i, j], 'd'),
                horizontalalignment="center",
                color="white" if cm[i, j] > thresh else "black")
    
    plt.tight_layout()
    plt.show()
    
    # Show some example predictions
    plt.figure(figsize=(15, 5))
    for i in range(5):
        plt.subplot(1, 5, i+1)
        plt.imshow(test_data[0][i])
        pred = predictions[i][0]
        true_label = test_data[1][i]
        color = 'green' if (pred > 0.5) == true_label else 'red'
        plt.title(f'Pred: {pred:.2f}\nTrue: {true_label}', color=color)
        plt.axis('off')
    plt.tight_layout()
    plt.show()
    
except FileNotFoundError:
    print("Model file not found. Make sure training completed successfully.")
except Exception as e:
    print(f"Error making predictions: {str(e)}")
````
